```python
import pandas as pd
import os

# 1. ระบุตำแหน่งไฟล์ (เปลี่ยนชื่อไฟล์ให้ตรงกับของคุณนะ)
file_path ="D:\\NaCl-Analysis-21X4\\Data\\data_Chem02_21X4.xlsx"

# 2. โหลดข้อมูล
df = pd.read_excel(file_path)

# 3. ดู 5 แถวแรกเพื่อเช็คหัวตาราง
print("--- 5 แถวแรกของข้อมูล ---")
display(df.head())

# 4. เช็คความพร้อมของข้อมูล (มี Missing value ไหม? ประเภทข้อมูลเป็นอะไร?)
print("\n--- ข้อมูลภาพรวม ---")
df.info()

# 5. สถิติเบื้องต้น (ดูค่า Min/Max ของอุณหภูมิว่าถึง 95°C จริงไหม)
print("\n--- สถิติพื้นฐาน ---")
display(df.describe())
```

```python
import pandas as pd
import os

# 1. สร้างลิสต์เพื่อเก็บข้อมูลที่จัดระเบียบแล้ว
clean_rows = []

# 2. วนลูปดึงข้อมูลทีละ 3 บรรทัด (RAW_Temp, M_Temp, EC)
for i in range(0, len(df), 3):
    try:
        row_temp   = df.iloc[i]
        row_m_temp = df.iloc[i+1]
        row_ec     = df.iloc[i+2]

        model_name = row_temp['model']
        nacl_val   = row_temp['NaCl_Percent']

        for rep_num in range(1, 36):
            clean_rows.append({
                'model':       model_name,
                'NaCl_Percent': nacl_val,
                'Rep':         rep_num,
                'RAW_Temp':    row_temp[rep_num],
                'M_Temp':      row_m_temp[rep_num],
                'EC':          row_ec[rep_num]
            })
    except Exception as e:
        print(f"Error at index {i}: {e}")

# 3. สร้าง DataFrame ใหม่
df_clean = pd.DataFrame(clean_rows)

# 4. เช็คจำนวนข้อมูลแยกตาม Model
print("--- จำนวนข้อมูลที่จัดระเบียบแล้ว ---")
print(df_clean['model'].value_counts())

# 5. ดูหน้าตาตารางใหม่
display(df_clean.head(10))

# ──────────────────────────────────────────────
# 6. บันทึกไฟล์แยกตาม Model
# ──────────────────────────────────────────────
SAVE_DIR = r"D:\NaCl-Analysis-21X4\Data"
os.makedirs(SAVE_DIR, exist_ok=True)          # สร้างโฟลเดอร์อัตโนมัติถ้ายังไม่มี

saved_files = []
for model_name, df_model in df_clean.groupby('model'):
    file_path = os.path.join(SAVE_DIR, f"{model_name}.csv")
    df_model.to_csv(file_path, index=False, encoding='utf-8-sig')  # utf-8-sig รองรับภาษาไทยใน Excel
    saved_files.append((model_name, len(df_model), file_path))
    print(f"✅ บันทึก {model_name}: {len(df_model)} แถว → {file_path}")

print(f"\n📁 บันทึกทั้งหมด {len(saved_files)} ไฟล์ ที่ {SAVE_DIR}")
```

```python
# นับจำนวนค่าที่ไม่ใช่ค่าว่างในคอลัมน์หลัก
total_values = df_clean[['RAW_Temp', 'M_Temp', 'EC']].notna().sum().sum()

print(f"จำนวนข้อมูลทั้งหมดที่บันทึกได้: {total_values} จาก 8,820")

if total_values == 8820:
    print("✅ ข้อมูลครบถ้วนสมบูรณ์ 100%!")
else:
    print(f"⚠️ มีข้อมูลหายไป {8820 - total_values} จุด")
```

```python
# ดูค่าสถิติแยกตาม Model เพื่อเช็คความผิดปกติ
summary = df_clean.groupby('model')[['RAW_Temp', 'EC']].agg(['min', 'max', 'mean']).round(2)

print("--- สรุปสถิติเปรียบเทียบ Hardware 4 รุ่น ---")
display(summary)

# แถม: เช็คค่า NaCl_Percent เพื่อยืนยันว่าทุกเครื่องทดสอบครบ 21 ระดับ
nacl_check = df_clean.groupby('model')['NaCl_Percent'].nunique()
print("\n--- จำนวนระดับ NaCl ที่แต่ละเครื่องทดสอบ (ควรเป็น 21) ---")
print(nacl_check)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ตั้งค่าฟอนต์และสไตล์
sns.set_theme(style="ticks")
plt.figure(figsize=(14, 8))

# พล็อตกราฟเปรียบเทียบทั้ง 4 Model
# เลือกดูเฉพาะระดับ NaCl สูงสุด (เช่น 2.0%) เพื่อให้เห็นความต่างชัดๆ หรือดูทั้งหมดก็ได้
sns.lineplot(data=df_clean, x='RAW_Temp', y='EC', hue='model', alpha=0.7)

plt.title('EC vs Temperature Comparison (Model 01-04)', fontsize=15)
plt.xlabel('Reference Temperature (°C)', fontsize=12)
plt.ylabel('Electrical Conductivity (EC)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Hardware Unit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# หาร 100 เพราะดูจากข้อมูลที่คุณส่งมา M_Temp น่าจะเป็นหลักพัน
sns.scatterplot(x=df_clean['RAW_Temp'], y=df_clean['M_Temp']/100, hue=df_clean['model'], s=10)
plt.plot([20, 100], [20, 100], color='red', linestyle='--') # เส้นอ้างอิง 1:1
plt.title('Sensor Temp vs Reference Temp')
plt.show()
```

```python
# 1. คำนวณค่าเฉลี่ย EC ของแต่ละ Model ในแต่ละระดับ NaCl
# เพื่อดูว่าที่ความเข้มข้นเดียวกัน เครื่องไหนอ่านค่าได้สูงหรือต่ำกว่าเพื่อน
pivot_comparison = df_clean.groupby(['NaCl_Percent', 'model'])['EC'].mean().unstack()

print("--- ค่าเฉลี่ย EC เปรียบเทียบ 4 Model (แยกตามระดับความเข้มข้น) ---")
display(pivot_comparison.head())

# 2. พล็อตกราฟแท่งเพื่อดูความต่างภาพรวม
import matplotlib.pyplot as plt

pivot_comparison.mean().plot(kind='bar', figsize=(10, 6), color=['#0F8B8D', '#EC9A29', '#A8201A', '#143642'])
plt.title('Average EC Reading per Model (All Concentrations)')
plt.ylabel('Mean EC Value')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
